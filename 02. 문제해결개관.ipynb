# 02. 문제 해결 개관
## 2-1. 파인만 알고리즘
1. 칠판에 문제를 적는다.
2. 골똘히 생각한다.
3. 칠판에 답안을 적는다.

## 2-2. 문제 해결 과정
1. 문제를 읽고 이해한다.
2. 문제를 익숙한 용어로 재정의한다.
3. 어떻게 해결할지 계획을 세운다. 
   - 어떤 자료구조 알고리즘을 사용할 것인가?
4. 계획을 검증한다.
5. 프로그램으로 구현한다.
6. 어떻게 풀었는지 되돌아보고, 개선할 방법이 있는지 찾아본다.
    - 코드와 자신의 접근법, 결정적인 깨달음, 못 푼 이유 등 기록.
