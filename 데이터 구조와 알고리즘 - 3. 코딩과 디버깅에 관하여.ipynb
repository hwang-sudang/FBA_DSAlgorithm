## 3.1 코딩의 중요성을 간과하지 말라

### [디버깅](https://ko.wikipedia.org/wiki/%EB%94%94%EB%B2%84%EA%B7%B8)

> **프로그래밍 단계 중에 발생하는 시스템의 논리적인 오류나 비정상적 연산(버그)을 찾아내고 그 원인을 밝히고 수정하는 작업 과정을 일컫는다. 일반적으로 디버깅을 하는 방법으로 테스트 상의 체크, 기계를 사용하는 테스트, 실제 데이터를 사용해 테스트하는 법이 있다.**


### 안티 디버깅

> **디버깅을 방지하고 분석하지 못하도록 하는 기술이다. 어떤 프로그램에 대해 디버깅이 가능하다는 것은 이 프로그램의 모든 코드를 확인할 수 있다는 것과 동치이다. 이를 통하여 개발자가 원하지 않았던 방식으로 처리 흐름을 바꾸거나 데이터를 바꿔 치기 하는 등 악용할 수 있는 소지가 많다. 따라서 만약 디버깅을 당한다면 해당 디버거 프로그램을 종료시키거나 에러를 발생시키는 방법 등 다양한 방법을 사용하여 분석을 방해한다.**

### 3.2.3 표준 라이브러리 공부하기
* 표준 라이브러리는 셀 수 없이 많이 사용되고 검증되었기 때문에, 메모리 관리나 정당성 증명에 신경을 쓸 필요 없이 편하게 사용가능.

* 특히 팀으로 참가하는 대회에서는 작성한 코드를 팀원들 역시 이해하기 쉬워야하므로 표준 라이브러리의 사용은 필수적.

### 3.2.4 항상 같은 형태로 프로그램을 작성하기
* 같은 코드를 다양한 방식으로 작성하는 것은 자신이 가장 알아보기 쉬운 코드를 작성하는 데에 도움이 되기 때문에 권장할 만하나, 시간이 지남에 따라 점점 실수의 원인이 되기도 함.

* 같은 코드를 작성하는 더 좋은 방법에 대하여 꾸준히 고민할 필요는 있지만, 자주 작성하는 알고리즘이나 코드 등에 대해서는 한 번 검증된 코드를 작성하고 이것만을 꾸준히 사용할 필요가 있음. 그래야 도구가 아닌 문제에 집중할 수 있기 때문!

#### strict weak ordering
1. **비반사성(Irreflexitivity)**: $a < a$ 는 항상 거짓.

2. **비대칭성(Asymmetry)**: $a < b$ 가 참이면 $b < a$는 거짓.

3. **전이성(Transitivity)**: $a < b \text{  and  } b < c \Rightarrow a < c$

4. **상등관계의 전이성(Transitivity of Equivalence)**: $a < b$ 와 $b < a$가 모두 거짓이면 $a = b$

## 3.4 디버깅과 테스팅

### 3.4.1 디버깅에 관하여

* 프로그래밍을 준비하는 사람은 디버거 없이 프로그램의 버그를 찾아내는 연습을 할 필요가 있음.

* 잘 분리된 기능적인 코드는 디버거 없이 눈만으로도 검증하기가 비교적 쉬움.

* **디버거를 사용하는 대신 거치면 좋은 과정**
 * 작은 입력에 대해 제대로 실행되나 획인하기
  
 * 단정문(assertion) 쓰기
  
 * 프로그램의 계산 중간 결과를 출력하면서 프로그래밍 하기

* 위의 과정들을 거쳐도 무엇이 틀렸는지 모를 경우에는 디버거를 사용해도 좋음.


### 3.4.2 테스트에 관하여

* 예제 입출력 외에도 몇 가지 간단한 입력을 직접 만들어 넣어보기

* 주어진 예제 입력을 약간 바꿔서 넣어 보거나, 있을 수 있는 가장 작은 입력과 가장 큰 입력을 넣어보기 

* **스캐폴딩(Scaffolding)  
 : 다른 코드를 개발할 때 뼈대를 잡기 위하여 임시로 사용하는 코드**

### 3.5.2 너무 큰 결과

### 3.5.3 너무 큰 중간 값
    
* 프로그램의 중간 과정에서 큰 값을 일시적으로 계산해야 하는 경우     
    

### 3.5.4 너무 큰 무한대 값

* 프로그램을 짜다 보면 무한대에 해당하는 큰 값을 이용하는 것이 편리할 때가 있음. 

* 해당 자료형이 처리 할수 있는 가장 큰 값을 무한대 값으로 쓰는 것, 예를 들어 32비트 부호 있는 정수라면 2^31 -1

* 무한대 값을 선택할 때에는 이 무한대 값들이 서로 더해지거나 곱해지는 경우가 없는지 잘 살펴보아야 함. 


### 3.5.5 오버플로 피해가기
* 가장 간단한 방법은 더 큰 자료형을 쓰는 것.

* 문제에 따라 계산 방법을 다르게 하여 오버플로를 피해가는 방법도 있음. (ex : 이항 계수)
